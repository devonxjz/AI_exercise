```python
ROWS = 3
COLS = 3

grid = [
    [1, 0, 1],
    [0, 1, 0],
    [1, 0, 1]
]

vacuum_pos = [1, 1]

# hướng quét
moving_right = True


def show_grid():

    print("\nMÔI TRƯỜNG:")

    for r in range(ROWS):
        for c in range(COLS):

            if [r, c] == vacuum_pos:
                print("V", end=" ")

            elif grid[r][c] == 1:
                print("1", end=" ")

            else:
                print("0", end=" ")

        print()

    print()


def is_dirty_exist():

    for r in range(ROWS):
        for c in range(COLS):

            if grid[r][c] == 1:
                return True

    return False


def suck():

    r = vacuum_pos[0]
    c = vacuum_pos[1]

    grid[r][c] = 0

    print(f"HÚT BỤI tại ({r},{c})")


def move(direction):

    if direction == "UP":
        vacuum_pos[0] -= 1

    elif direction == "DOWN":
        vacuum_pos[0] += 1

    elif direction == "LEFT":
        vacuum_pos[1] -= 1

    elif direction == "RIGHT":
        vacuum_pos[1] += 1

    print(f"DI CHUYỂN {direction} -> {vacuum_pos}")

def state():

    global moving_right

    r = vacuum_pos[0]
    c = vacuum_pos[1]

    # RULE 1: hút bụi
    if grid[r][c] == 1:
        suck()
        return

    # QUÉT ZIGZAG
    if moving_right:

        if c < COLS - 1:
            move("RIGHT")

        elif r < ROWS - 1:
            move("DOWN")
            moving_right = False

    else:

        if c > 0:
            move("LEFT")

        elif r < ROWS - 1:
            move("DOWN")
            moving_right = True


def main():

    step = 0

    print("MÁY HÚT BỤI AI\n")

    while is_dirty_exist():

        step += 1

        print(f"--- BƯỚC {step} ---")

        show_grid()

        state()

    print("\nHOÀN THÀNH!")
    show_grid()


main()
```

```python
from enum import Enum

class Action(Enum):
    UP = "UP"
    DOWN = "DOWN"
    LEFT = "LEFT"
    RIGHT = "RIGHT"
    NONE = "NONE"

class Rule:
    def __init__(self, row, col, actions):
        self.row = row
        self.col = col
        self.actions = actions

class State:
    def __init__(self, board):
        self.board = board

    def print_board(self):
        for row in self.board:
            print(row)

RULES = []

for i in range(3):
    for j in range(3):
        actions = []
        if i > 0:
            actions.append(Action.UP)
        if i < 2:
            actions.append(Action.DOWN)
        if j > 0:
            actions.append(Action.LEFT)
        if j < 2:
            actions.append(Action.RIGHT)
        RULES.append(Rule(i, j, actions))

def interpret_input(percept):
    for i in range(3):
        for j in range(3):
            if percept.board[i][j] == 0:
                return (i, j)
    return None

def rules_match(row, col):
    for rule in RULES:
        if rule.row == row and rule.col == col:
            return rule
    return None

def action(rule):
    if rule is None or len(rule.actions) == 0:
        return Action.NONE
    return rule.actions[0]

def simple_reflex_agent(percept):
    row, col = interpret_input(percept)

    rule = rules_match(row, col)

    return action(rule)

board = []

print("Nhập ma trận 8 puzzle:")

for i in range(3):
    row = list(map(int, input(f"Hàng {i+1}: ").split()))
    board.append(row)

state = State(board)

print("\nCurr State:")
state.print_board()

selected_action = simple_reflex_agent(state)

print("\nAction được chọn:")
print(selected_action.value)
```

```python
from enum import Enum

class Action(Enum):
    UP = "UP"
    DOWN = "DOWN"
    LEFT = "LEFT"
    RIGHT = "RIGHT"
    NONE = "NONE"

class State:

    def __init__(self, board):
        self.board = board

    def print_board(self):

        for row in self.board:
            print(row)

    def find_blank(self):

        for i in range(3):
            for j in range(3):

                if self.board[i][j] == 0:
                    return (i, j)

        return None

class ModelBasedAgent:

    def __init__(self):

        self.model = None

    def update_model(self, percept):

        self.model = percept

    def possible_actions(self):

        row, col = self.model.find_blank()

        actions = []

        if row > 0:
            actions.append(Action.UP)

        if row < 2:
            actions.append(Action.DOWN)

        if col > 0:
            actions.append(Action.LEFT)

        if col < 2:
            actions.append(Action.RIGHT)

        return actions

    def choose_action(self):

        actions = self.possible_actions()

        if len(actions) == 0:
            return Action.NONE

        priority = [
            Action.UP,
            Action.LEFT,
            Action.DOWN,
            Action.RIGHT
        ]

        for p in priority:
            if p in actions:
                return p

        return Action.NONE

    def move(self, action):

        row, col = self.model.find_blank()

        new_row = row
        new_col = col

        if action == Action.UP:
            new_row -= 1

        elif action == Action.DOWN:
            new_row += 1

        elif action == Action.LEFT:
            new_col -= 1

        elif action == Action.RIGHT:
            new_col += 1

        self.model.board[row][col], self.model.board[new_row][new_col] = \
            self.model.board[new_row][new_col], self.model.board[row][col]

board = []

print("Nhập ma trận 8 puzzle:")

for i in range(3):

    row = list(map(int, input(f"Hàng {i+1}: ").split()))
    board.append(row)

state = State(board)

agent = ModelBasedAgent()

# percept -> model
agent.update_model(state)

print("\nTRẠNG THÁI HIỆN TẠI:")
agent.model.print_board()

# chọn action
selected_action = agent.choose_action()

print("\nACTION ĐƯỢC CHỌN:")
print(selected_action.value)

# thực hiện action
agent.move(selected_action)

print("\nTRẠNG THÁI SAU KHI DI CHUYỂN:")
agent.model.print_board()
```

```python
from enum import Enum

class Action(Enum):
    UP = "UP"
    DOWN = "DOWN"
    LEFT = "LEFT"
    RIGHT = "RIGHT"
    SUCK = "SUCK"
    NONE = "NONE"

class Environment:

    def __init__(self, grid):

        self.grid = grid

    def print_grid(self, vacuum_pos):

        print("\nMÔI TRƯỜNG:")

        for i in range(3):

            for j in range(3):

                if [i, j] == vacuum_pos:
                    print("V", end=" ")

                else:
                    print(self.grid[i][j], end=" ")

            print()

        print()

    def is_dirty_exist(self):

        for row in self.grid:
            for cell in row:

                if cell == 1:
                    return True

        return False

class ModelBasedVacuumAgent:

    def __init__(self):

        # model lưu trạng thái môi trường
        self.model = None

        # vị trí máy hút bụi
        self.position = [0, 0]

        # hướng quét zigzag
        self.moving_right = True

    # cập nhật model
    def update_model(self, percept):

        self.model = percept

    # cảm biến
    def perceive(self):

        r = self.position[0]
        c = self.position[1]

        return self.model.grid[r][c]

    # hút bụi
    def suck(self):

        r = self.position[0]
        c = self.position[1]

        self.model.grid[r][c] = 0

        print(f"HÚT BỤI tại ({r},{c})")

    # di chuyển
    def move(self, action):

        if action == Action.UP:
            self.position[0] -= 1

        elif action == Action.DOWN:
            self.position[0] += 1

        elif action == Action.LEFT:
            self.position[1] -= 1

        elif action == Action.RIGHT:
            self.position[1] += 1

        print(f"DI CHUYỂN {action.value} -> {self.position}")

    # chọn action dựa trên MODEL
    def choose_action(self):

        r = self.position[0]
        c = self.position[1]

        # RULE 1: nếu ô hiện tại bẩn -> hút
        if self.perceive() == 1:
            return Action.SUCK

        # RULE 2: quét zigzag
        if self.moving_right:

            if c < 2:
                return Action.RIGHT

            elif r < 2:
                self.moving_right = False
                return Action.DOWN

        else:

            if c > 0:
                return Action.LEFT

            elif r < 2:
                self.moving_right = True
                return Action.DOWN

        return Action.NONE

    # thực hiện action
    def act(self, action):

        if action == Action.SUCK:
            self.suck()

        elif action != Action.NONE:
            self.move(action)


# =========================
# MAIN
# =========================

grid = [
    [1, 0, 1],
    [0, 1, 0],
    [1, 0, 1]
]

environment = Environment(grid)

agent = ModelBasedVacuumAgent()

# percept -> model
agent.update_model(environment)

step = 0

print("MODEL BASED VACUUM AGENT\n")

while environment.is_dirty_exist():

    step += 1

    print(f"--- BƯỚC {step} ---")

    environment.print_grid(agent.position)

    action = agent.choose_action()

    print("ACTION:", action.value)

    agent.act(action)

print("\nHOÀN THÀNH!")

environment.print_grid(agent.position)
```
